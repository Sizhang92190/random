```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lamb = 45   # 45 lambdas

#loading xvg file
data = np.loadtxt('data/dhdl_%dlambdas.txt'%lamb)
print ('data.shape=', data.shape)

dhdl = data[:, 4:]   ##only loading # ensembles
print ('dhdl.shape=', dhdl.shape)

#loading sampled lambda indices
thermo_states = data[:, 1]
print ('thermo_states=', thermo_states)
```

```python
###Plotting the distribution of the delta_energies
plt.figure(figsize=(8,30))
for i in range(lamb-1):   
    Ind = (thermo_states == i)
    
    delta_u = dhdl[Ind, i+1]
    #print (delta_u.max())
    #print (len(delta_u))   
    
    delta_u_bins = np.arange(-100., 100., 1.)
    counts, bin_edges = np.histogram(delta_u, bins=delta_u_bins)
    bin_centers = (bin_edges[0:-1] + bin_edges[1:])/2.0
    
    plt.subplot(lamb, 1, i+1)
    plt.plot(bin_centers, counts)
    plt.xlabel('$\Delta u_{%d \\rightarrow %d}$'%(i, i+1))
#plt.tight_layout()
plt.show()
```
