```python
import numpy as np
matrix = np.load('transition_matrix.npy')    #for a bad toy model
print (matrix[:, 0].sum())

#matrix = matrix*100
print (matrix.shape)
#print (matrix)

n_ensembles = len(matrix)
print ('n_ensembles=', n_ensembles)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline  

# plotting transition matrix

plt.figure(figsize=(4,3))
plt.pcolor(matrix, cmap='binary')
plt.xlabel('thermodynamic index', fontsize='large')
plt.ylabel('thermodynamic index', fontsize='large')
plt.tick_params(axis='both', labelsize='large')
plt.tight_layout()
plt.show()
```

```python
#---- calculating tao(lambda1 -> lambdaM), forward direction
M=n_ensembles-1

tao_forward=0
for i in range(0, M):
    for j in range(0, i+1):     #i+1, including i         
            term_1 = 1/matrix[j, j+1]
        
            Y=[]  
            for k in range(j+1, i+1):     #i+1, including i 
                Y.append(matrix[k, k-1]/matrix[k, k+1])            
            term_2 = np.prod(Y)
        
            new = term_1 * term_2
            tao_forward += new
print ('tao_forward=', tao_forward)


#---- calculating tao(lambdaM -> lambda1), backword direction
M=n_ensembles

tao_backword=0
for i in range(1, M):   
    for j in range(i, M): 
        term_1 = 1/matrix[j, j-1]       
        
        Y=[]
        for k in range(i, j):               
            Y.append(matrix[k, k+1]/matrix[k, k-1])
            
        term_2 = np.prod(Y)
        
        new = term_1 * term_2
        tao_backword += new
print ('tao_backword=', tao_backword)

print ('mean round trip time=', tao_forward+tao_backword, 'in unit of steps')
```
